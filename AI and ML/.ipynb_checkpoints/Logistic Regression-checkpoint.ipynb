# Logistic Regression

- Can be used to fit multiple features using non linear functions
- A regression model that is used in binary classification

## The model
- Does linear regression
- Uses sigmoid activation function 
- Maps the predictions of linear regression to the activation function
- Produces an output between 0 and 1

```python
# Required imports
import math
import numpy as np
import matplotlib.pyplot as plt
```

## Creating the Sigmoid Function

```python
def sigmoid_val(x):
    return 1 / (1 + math.e ** -(x))

print(sigmoid_val(100))
print(sigmoid_val(-100))
```

## Visualising the Sigmoid Function

```python
x_val = np.arange(-20, 21, 1)
y_val = np.array([sigmoid_val(i) for i in x_val])

# Plotting the graph
plt.plot(x_val, y_val, c = "b", label = "Sigmoid Value")

# Adding the legend
plt.title("Sigmoid Function")
plt.xlabel("Array Value")
plt.ylabel("Sigmoid Value")
plt.legend()

# Displaying the graph
plt.show()
```
