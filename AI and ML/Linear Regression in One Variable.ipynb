# Creating a Linear Regression Model from scratch to predict the price of houses based on the training data

```python
# Importing the essential packages
import numpy as np
import matplotlib.pyplot as plt
```

## Creating the training data

- `x_train`: Features used in training
- `y_train`: Labels used in training

```python
# Feature: Size in 1000 sqft
x_train = np.array([1, 2], dtype = "float32")

# Target: Price in 1000 dollars
y_train = np.array([300, 500], dtype = "float32")

print(f"Size of Houses: {x_train}")
print(f"Prize of Houses: {y_train}")
```

```python
# Finding the no of elements in the training data 'm'
print(f"The no of examples in the training data: {x_train.shape[0]}")
```

```python
# Accessing the ith Training example and its target
print(f"(x ^ (0), y ^ (0)) = ({x_train[0]}, {y_train[0]})")
```

## Plotting the training data

```python
# Plotting the scatter plot
plt.scatter(x_train, y_train, marker = ".", c = "b")

# Adding legends to the plot
plt.title("Housing Prices")
plt.xlabel("House Area (in Sqft)")
plt.ylabel("House Price (in Dollars)")

# Displaying the plot
plt.show()
```

## Constructing the Regression Model

```python
# Taking Initial values of model parameters
w, b = 200, 100

print(f"w = {w}")
print(f"b = {b}")
```

## Working of a Linear Regression Model

- `f(x) = w[i] * x[i] + b`

- Where `w[i]`, `x[i]` iterates over all the values in the training data

- For each evaluation a prediction ÿ is made

```python
def compute_model_output(x, w, b):
    
    """
    Computes the prediction for each value of the training dataset
    Arguments:
        x - Features
        w, b - Model Parameters
        
    Returns:
        y - Target values
    """
    
    # Counting the total no of examples
    m = x.shape[0]
    
    # Creating an array to store all the predictions
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

## Using the compute_model_output() to find all the necessary predictions

```python
# Storing the predictions
pred_f_wb = compute_model_output(x_train, w, b)

# Plotting them against the actual values
plt.plot(x_train, pred_f_wb, c = "b", label = "Predicted Values")
plt.scatter(x_train, y_train, c = "r", marker = "o", label = "Actual Values")

# Adding legends to the plot
plt.title("Housing Prices")
plt.xlabel("House Area (in Sqft)")
plt.ylabel("House Price (in Dollars)")
plt.legend()

# Displaying the plot
plt.show()
```

# Trial and Error

- After trying many random values of w and b
- w = 200 and b = 100 perfectly fits the training set

# Prediction

Predicting the cost of a house on an un - trained value

```python
# Size of house is 1200 sqft
x_i = 1.2

# Using the parameters after tuning
f_xi = w * x_i + b

print(f"The cost of a house of 1200 sqft is {f_xi} thousand dollars")
```
