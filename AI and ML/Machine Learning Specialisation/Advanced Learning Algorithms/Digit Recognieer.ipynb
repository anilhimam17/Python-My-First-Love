```python
# Necessary Imports
import tensorflow as tf
```

```python
# Loading the Dataset
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# Checking the features and labels
print(f"Features: {digits.feature_names}")
print(f"Features: {digits.target_names}")
```

```python
# Checking the shape of the data
print(f"No of examples: {digits.data.shape}")
```

```python
# Checking the charecteristrics of a single training example
print("Features: ", digits.data[0])
print("Target: ", digits.target[0])
```

```python
# Loading the training and testing data
X_train = digits.data[:1700]
y_train = digits.target[:1700]

X_test = digits.data[1700:]
y_test = digits.target[1700:]
```

```python
# Designing the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 8, activation = "relu"),
    tf.keras.layers.Dense(units = 8, activation = "relu"),
    tf.keras.layers.Dense(units = 10, activation = "relu"),
    tf.keras.layers.Dense(units = 10, activation = "softmax")
])

# Compiling the model
model.compile(loss = "SparseCategoricalCrossentropy", optimizer = "adam")

# Training the model
model.fit(X_train, y_train, epochs = 100)
```

```python
# Predicting using the model
model.predict(X_test)
```
