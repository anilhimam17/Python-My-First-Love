```python
import matplotlib.pyplot as plt
import numpy as np
```

## Working on Basic Plots - Functional Approach

```python
# Array of values to be plotted on the graph
x_val = np.linspace(1, 20, 50)
y_val = x_val ** 2
```

## Plotting a Line chart

```python
plt.plot(x_val, y_val)
plt.show()
```

## Subplots - Functional Approach

- plt.subplots(`row`, `column`, `no of graph`)

```python
# Plotting the first subplot
plt.subplot(1, 2, 1)
plt.plot(x_val, y_val)

# Plotting the second subplot
plt.subplot(1, 2, 2)
plt.plot(y_val, x_val)

# Displaying the graph
plt.show()
```

## Using the object oriented approach

In the object oriented approach a figure object similar to a blank canvas is created and 
everything is built on to the canvas.

```python
# Creating the figure object
fig = plt.figure()

# Adding axes to the figure
axes = fig.add_axes([0.1, 0.1, 1.0, 0.8])

# Plotting on the custom axes
axes.plot(x_val, y_val)

# Anatomy of the chart
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("Object Oriented Approach")
```

## Automating the Subplots using OOP

- fig, axes = plt.subplots(`nrows`, `ncols`)

- This automatically returns an array of plot axes of equal measures to plot different charts

```python
# Creating an automated set of axes
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Using automated set of axes to plot subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plotting graphs on the different axes
axes[0].plot(x_val, y_val)
axes[1].plot(y_val, x_val)
plt.tight_layout()
```

## Saving a figure using matplotlib

- plt.savefig(`Filename / Path.Extension`)

```python
# Using automated set of axes to plot subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plotting graphs on the different axes
axes[0].plot(x_val, y_val)
axes[1].plot(y_val, x_val)
plt.tight_layout()

plt.savefig("Subplots.png", dpi = 200)
```

## Adding a Legend

- axes.legend(`location`)

- Where the location attribute is used to specify the location to place the legend on the graph

```python
# Creating the figure object
fig = plt.figure()

# Adding the axes to the figure
axes = fig.add_axes([0.1, 0.1, 1.0, 1.0])

# Plotting the data on the axes
axes.plot(x_val, x_val ** 0.5, label = "Square Root of X")
axes.plot(x_val, np.log(x_val), label = "Log of X")

# Adding the legend for the graph
# axes.legend(loc = 0)
# axes.legend(loc = "lower right")
axes.legend(loc = (0.6, 0.1))
```
