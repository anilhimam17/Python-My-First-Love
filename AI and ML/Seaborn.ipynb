## Seaborn

### It is a statistical data visualisation library which provides several mathematical distribution functions for better visualisation of the data

```python
# Importing the library
import seaborn as sns
```

```python
# Importing a dataset from seaborn to visualise
tips = sns.load_dataset("tips")
```

```python
# Exploring the dataframe
print(tips.head())
```

```python
print(tips.describe())
```

```python
print(tips.info())
```

## Distribution Plots

- A way to visualise one of more columns in a dataframe using mathematical visualisation and comparison techniques

## Distplot

- A method to visualise univariate data using a kernal density estimation of all the available values

- sns.distplot(`column`)

```python
sns.displot(tips["total_bill"], kde = True)
```

## Jointplots

- Jointplots allow the comparision of distplots of two columns. Hence they allow the comparison of bivariate data

- sns.jointplot(`x - column`, `y - column`, `dataset`)

- Jointplots can take a parameter `kind` to change the type of graph being used to compare the columns

- Jointplots can take a parameter `hue` to view categorical features on a graph while comparing the columns

```python
# Visualising the relation between tip amt and bill amt
sns.jointplot(x = tips["total_bill"], y = tips["tip"], data = tips, kind = "hex")
```

```python
sns.jointplot(x = tips["total_bill"], y = tips["tip"], data = tips, hue = tips["sex"])
```

## Pairplots

- Pairplots are provided by Seaborn to compare multivarite data
- Pairplots compare every single column available in the dataset with eachother

- sns.pairplot(`dataset`)

```python
sns.pairplot(tips)
```

## Categorical Plots

- Categorrical plots in Seaborn are used to visualise the distribution of a categorical column wrt any numerical column.

- This allows checking the performance of each numerical column using the categories available

## Barplots

- They are used to visualise the distribution of a categorical column on the basis of any estimator

- sns.barplot(`x - Categorical Column`, `y - Numerical Column`, `data`, `estimator`)

```python
sns.barplot(x = tips["sex"], y = tips["total_bill"], estimator = "max")
```

## Countplots

- They are a special case of barplots whose estimator only checks for the number values available in the different categories.

- They only check the performance of all the categories in a single categorical column

- sns.countplot(`x`, `data`)

```python
sns.countplot(x = tips["day"], data = tips)
```

## Boxplots and Violinplots

- They are used to visualise the distribution of any categorical columns

- sns.boxplot(`x`, `y`, `data`)

- sns.violinplot(`x`, `y`, `data`)

```python
sns.boxplot(x = tips["day"], y = tips["total_bill"], data = tips)
```

```python
sns.boxplot(x = tips["day"], y = tips["total_bill"], data = tips, hue = "sex")
```

```python
sns.violinplot(x = tips["day"], y = tips["tip"], data = tips)
```

```python
sns.violinplot(x = tips["day"], y = tips["tip"], data = tips, hue = "sex", split = True)
```

## Stripplots

- They are provided by Seaborn to get a scatter plot like visualisation of the distribution of categorical data.

- Stripplots take a parameter * `jitter` = `true` * to get a `2D` visualisation of the distribution

- sns.stripplot(`x`, `y`, `data`)

```python
sns.stripplot(x = tips["day"], y = tips["tip"], data = tips)
```

```python
sns.stripplot(x = tips["day"], y = tips["tip"], data = tips, hue = "sex")
```

## Swarmplots

- They are a combination of violin plots and stripplots
- Swarmplots can be overlayed on violin plots to get a better visualisation for the KDE of the violin plots

- sns.swarmplot(`x`, `y`, `data`)

```python
sns.swarmplot(x = tips["day"], y = tips["total_bill"], data = tips, hue = "sex")
```

```python
# Overlaying a swarmplot on a violinplot
sns.violinplot(x = tips["day"], y = tips["total_bill"], data = tips)
sns.swarmplot(x = tips["day"], y = tips["total_bill"], data = tips, color = "blue")
```

## Matrix Plots

- Matrix plots are provided by seaborn to visualise the entire dataframe if the information is in the form a matrix.

- There are two primary matrix plots provided by seaborn: Heatmaps and Clustermaps

## Heatmap
- sns.heatmap(`matrix data`, `annot`, `cmap`, `linewidth`, `linecolor`)

```python
# Loading the matrix data
flights = sns.load_dataset("flights")
fp = flights.pivot_table(index = "month", columns = "year", values = "passengers")

# Creating a corelation matrix
corr_tips = tips.corr()
```

```python
print(corr_tips)
```

```python
sns.heatmap(corr_tips, annot = True)
```

```python
print(fp)
```

```python
sns.heatmap(fp, linewidths = 0.5, cmap = "coolwarm")
```

## Clustermap
- A clustermap is an intelligent version of a heatmap which groups similar data together automatically
- sns.clustermap(`matrix data`, `annot`, `cmap`, `linewidth`, `linecolor`)

```python
sns.clustermap(fp, cmap = "coolwarm")
```

## Grids

- Seaborn allows programmer defined grids for more flexibility in the functionality of a the `pairplot()` and the `jointplot()`

- Seaborn provides two types of grids
- Facet Grids()
- Pair Grids()

## Seaborn Facet Grids

```python
grid = sns.FacetGrid(data = tips, col = "day", row = "smoker")
grid.map(sns.distplot, "total_bill")
```

## Seaborn Pair Grids

```python
grid = sns.PairGrid(tips)
grid.map_diag(sns.swarmplot)
grid.map_upper(sns.barplot)
grid.map_lower(sns.kdeplot)
```

## Regression Plots

- Lastly seaborn provides regression plots to fit linear models to any given data

```python
sns.lmplot(x = "tip", y = "total_bill", data = tips, hue = "sex")
```

### Regression plots have a additional functionality where they can take extra parameters to be 
### turned into a `Facet Grid` like group of plots

```python
sns.lmplot(x = "tip", y = "total_bill", data = tips, col = "sex", row = "day")
```
